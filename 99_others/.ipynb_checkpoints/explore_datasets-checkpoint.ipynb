```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
dataset_path = "../00_datasets/dataset_output/libras_minds/libras_minds_openpose.csv"
```

```python
df = pd.read_csv(dataset_path)
```

```python
df_video = df[df["video_name"] == "01AcontecerSinalizador01-1.mp4"]
```

```python
df_video
```

```python
df_video = df_video.drop(["Unnamed: 0", "category", "video_name", "missing_hand", "missing_face", "frame"], axis=1)
```

```python
df_video
```

```python
x_columns = [i for i in df_video.columns if i.endswith("x")]
y_columns = [i for i in df_video.columns if i.endswith("y")]
```

```python
type(df_video)
```

```python
x_values = df_video[x_columns].T.to_numpy()
y_values = df_video[y_columns].T.to_numpy()
```

```python
x_values
```

```python
x_values[x_values < 0] = 0
x_values[x_values > 1] = 1
```

```python
y_values[y_values < 0] = 0
y_values[y_values > 1] = 1
```

```python
x_values.min(), x_values.max()
```

```python
y_values.min(), y_values.max()
```

```python
x = x_values
y = y_values
channels = 3
```

```python
x.shape
```

```python
width = x.shape[1]
width
```

```python
# Leve ajuste para o tamanho da coluna ser múltiplo da quantidade de canais (RGB)
if width % channels != 0:
    extra_cols = width % channels
    x = x[:, : width - extra_cols]
    y = y[:, : width - extra_cols]
```

```python
x
```

```python
x.shape
```

```python
# -1 no reshape significa que ele vai encontrar um valor automático para ele
# Seria o mesmo que:
# np.reshape(x, (x.shape[0], int(x.shape[1] / channels), channels)).shape
X = np.reshape(x, (x.shape[0], -1, channels))
Y = np.reshape(y, (y.shape[0], -1, channels))
```

```python
# Novo formato (shape):
# Altura x Largura x Canais
X.shape
```

```python
X
```

```python
pd.DataFrame.from_records(X)
```

```python
# Junta os arrays pelo eixo 1 (coluna)
image = np.concatenate([X, Y], axis=1)
```

```python
image
```
