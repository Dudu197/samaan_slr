```python
import pandas as pd
import os
import json
```

```python
output_file = "models.csv"
path = "model_results"
file_name = "model_result.json"
```

```python
df = None
for model in os.listdir(path):
    with open(os.path.join(path, model, file_name), "r") as f:
        model_result = json.load(f)
    model_result["optimizer_learning_rate"] = model_result["optimizer"]["learning_rate"]
    model_result["optimizer_weight_decay"] = model_result["optimizer"]["weight_decay"]
    model_result["optimizer"] = model_result["optimizer"]["name"]
    model_result.pop('history', None)
    model_result.pop('summary', None)
    model_df = pd.DataFrame.from_dict([model_result])
    if df is None:
        df = model_df
    else:
        df = pd.concat([df, model_df], ignore_index=True)
df
```

```python
df.to_csv(output_file)
```

```python
df.to_excel(output_file.replace(".csv", ".xlsx"))
```
