```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
train_data = df = pd.read_csv("minds_train_dataset.csv")
test_data = df = pd.read_csv("minds_test_dataset.csv")
```

```python
frames = 150
```

```python
train_data.columns
```

```python
train_features = train_data.drop(["Unnamed: 0", "category", "video_name", "frame"], axis=1).values
train_target = train_data['category'].values
```

```python
test_features = test_data.drop(["Unnamed: 0", "category", "video_name", "frame"], axis=1).values
test_target = test_data['category'].values
```

```python
train_features = train_features.reshape((int(train_features.shape[0]/frames), frames, train_features.shape[1]))
train_target = train_target.reshape((int(train_target.shape[0]/frames), frames))[:, 0]
```

```python
test_features = test_features.reshape((int(test_features.shape[0]/frames), frames, test_features.shape[1]))
test_target = test_target.reshape((int(test_target.shape[0]/frames), frames))[:, 0]
```

```python
train_features.shape
```

```python
train_target.shape
```

```python
model = Sequential()
model.add(LSTM(64, input_shape=(frames, train_features.shape[2]), return_sequences=True))
model.add(LSTM(128, return_sequences=True))
model.add(LSTM(64))
model.add(Dense(64))
model.add(Dense(32))
model.add(Dense(20, activation='softmax'))
```

```python
# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='SGD', metrics=['accuracy'])
```

```python
# Train the model
model.fit(train_features, train_target, epochs=1000, batch_size=128)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(test_features, test_target)
print('Test loss:', loss)
print('Test accuracy:', accuracy)
```
